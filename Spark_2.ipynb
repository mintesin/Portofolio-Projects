```
from pyspark.sql import SparkSession
```

```
spark=SparkSession.builder.appName('Dataframe').getOrCreate()
dat=spark.read.option('header','True').csv('feature_matrix.csv',inferSchema=True)
```

```
dat.show()
```

```
dat.describe().show()
```

```
dat.na.drop(how="any").show()
```

```
dat.na.drop(how="any",subset=['latitude']).show()
```

```
#filling the missing value
dat.na.fill(28,'Latitude').show(10)
```

```
#imputation with median inplace of the null values

from pyspark.ml.feature import Imputer
cols=dat.columns
imputer = Imputer(
    inputCols=dat.columns, 
    outputCols=["{}_imputed".format(c) for c in cols]
    ).setStrategy("median")
```

```
imputer.fit(dat).transform(dat).show(10)
```

```
#filtering operations
dat.filter("Speed<=17").select("Speed","Latitude","Longitude").show(10)
```

```
dat.filter(dat['Speed']<=17).select("Speed","Latitude",'Longitude').show(10)
```

```
#multiple conditions
dat.filter((dat["Speed"]<=17) &
        (dat['PCI']==64)).select("Speed","Longitude","Latitude","PCI").show(10)
```

```
#Goupby to find the maximum speed 
dat.groupby('PCI').max('Speed').show()
```

```
dat.groupBy('PCI').avg("Speed").show()
```
